## ANALYSIS ON CRIME AGAINST WOMEN DATA

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("crimes_against_women_2001-2014.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

# Query 1
Correlation in Rape and Kidnapping in different states.

```python
selected_states = ["BIHAR", "KERALA", "UTTAR PRADESH", "ASSAM", "HIMACHAL PRADESH", "RAJASTHAN", "DELHI"]

selected_data = df[df["STATE/UT"].isin(selected_states)]
```

```python
sns.scatterplot(data = selected_data, x="Kidnapping and Abduction", y="Rape", hue="STATE/UT")
```

## <span style="color: red;">Conclusion:-</span>

This correlation between rape and kidnapping cases in Assam, Bihar, Gujarat, Himachal Pradesh, Kerala, Rajasthan, and Uttar Pradesh shows that Bihar has a high number of kidnapping cases, Rajasthan has a high number of rape cases, and Uttar Pradesh has both a high number of kidnapping and rape cases. After seeing this scatterplot, we can conclude that in these states, law and order is not great, and government policies are not effectively addressing the issue of crimes against women.

# Query 2
Correlation between crimes.

```python
col=["Rape", "Kidnapping and Abduction", "Dowry Deaths", "Cruelty by Husband or his Relatives", "Importation of Girls"]
sns.heatmap(df[col].corr(),cmap="coolwarm", annot=True)
```

## <span style="color: red;">Conclusion:-</span>

This heatmap shows us the correlation among various crimes, revealing a strong correlation between rape cases, dowry deaths, and kidnapping.

# Query 3
State wise comparison of total Rape Cases from 2001 to 2014.

```python
state_counts = selected_data.groupby("STATE/UT")["Rape"].sum()

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c2f0c2']  # Custom colors
explode = (0, 0, 0, 0, 0, 0, 0.1)  # Explode the first slice (Category A)

plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(state_counts, labels=state_counts.index, autopct='%1.1f%%', startangle=140,colors=colors, explode=explode, shadow=True)  # autopct adds percentage labels


plt.title('Distribution of Categories')


plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.legend(loc='upper right', labels=state_counts.index)
plt.show()
```

## <span style="color: red;">Conclusion:-</span>

This pie chart displays rape cases among Assam, Rajasthan, Bihar, Uttar Pradesh, Delhi, Kerala, and Himachal Pradesh, and it's evident that Uttar Pradesh has the highest number of cases. This signifies that law and order in Uttar Pradesh is inadequate, and in terms of women's safety, the state is not performing well.

## Query 5
State wise comparison of total Dowry Deaths Cases from 2001 to 2014.

```python
selected_states1 = ["BIHAR", "KERALA", "UTTAR PRADESH", "HIMACHAL PRADESH"]

selected_data1 = df[df["STATE/UT"].isin(selected_states1)]

sns.barplot(selected_data1, x="STATE/UT", y="Dowry Deaths")
```

## <span style="color: red;">Conclusion:-</span>

Deaths due to the dowry system have been recorded in Bihar, Himachal Pradesh, Kerala, and Uttar Pradesh. Among these states, Bihar and Uttar Pradesh have the highest numbers in India, while Himachal Pradesh and Kerala have the lowest. This disparity may be attributed to cultural ignorance or low literacy rates among the people in Bihar and Uttar Pradesh, and conversely, the higher literacy rate in Himachal Pradesh and Kerala.
